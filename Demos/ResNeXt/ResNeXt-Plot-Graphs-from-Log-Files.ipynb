# Use this Notebook to Plot Training and testing errors per epoch

```python
import numpy as np
import matplotlib.pyplot as plt
import json
import os
```

## Load the log file

```python
log_filename = 'resnext-0-log.txt'
log_dir= os.path.join(os.path.realpath(os.path.join(os.getcwd(), os.path.pardir, os. path.pardir)), 'log')
log_filepath = os.path.join(log_dir, log_filename)
print("Loading Logfile: {}".format(log_filepath))

with open(log_filepath) as f:
    lines = f.readlines()

epochs = []
test_loss = []
train_loss = []
test_acc = []

for line in lines:
    data = json.loads(line)
    epochs.append(data['epoch'])
    test_loss.append(data['test_loss'])
    train_loss.append(data['train_loss'])
    test_acc.append(data['test_accuracy'])
```

## Relevant Statisitics

```python
print("Maximum Test Accuracy (Minimum Test Error) = {}% ( {}% )".format(np.max(test_acc)*100, (1-np.max(test_acc))*100))
print("Minimum Training loss = {}".format(np.min(train_loss)))
print("Minimum Testing loss = {}".format(np.min(test_loss)))
```

## Plot The Testing and Training Loss per epoch

```python
plt.figure(figsize = (16,6))
plt.subplot(1,1,1)
plt.title('Training and Testing Loss per epoch for ResNeXt (Baseline) (150 epochs)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
resnext_train_loss_h = plt.plot(epochs, train_loss, label='Training Loss')
resnext_test_loss_h = plt.plot(epochs, test_loss, label='Testing Loss')
plt.legend(handles = [resnext_train_loss_h[0],resnext_test_loss_h[0]])
plt.show()
```

## Plotting the Testing Accuracy per epoch

```python
plt.figure(figsize = (16,4))
plt.subplot(1,1,1)
plt.title('Testing Accuracy per epoch for ResNeXt (Baseline) (150 epochs)')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
resnext_train_loss_h = plt.plot(epochs, np.array(test_acc)*100, label='Testing Accuracy')
plt.legend(handles = [resnext_train_loss_h[0]])
plt.show()
```
