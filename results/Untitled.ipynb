```python
import numpy as np
import matplotlib.pyplot as plt
import os

dn_train = np.load(os.path.abspath(os.path.join(os.getcwd(), 'dn_train_dbpn.npy')))
dn_test = np.load(os.path.abspath(os.path.join(os.getcwd(), 'dn_test_dbpn.npy')))
dn_train_dbpn_x2 = np.load(os.path.abspath(os.path.join(os.getcwd(), 'dn_train_x2_dbpn.npy')))
dn_test_dbpn_x2 = np.load(os.path.abspath(os.path.join(os.getcwd(), 'dn_test_x2_dbpn.npy')))
dn_train_dbpn_x4 = np.load(os.path.abspath(os.path.join(os.getcwd(), 'dn_train_x4_dbpn.npy')))
dn_test_dbpn_x4 = np.load(os.path.abspath(os.path.join(os.getcwd(), 'dn_test_x4_dbpn.npy')))

dn_train_bi_x2 = np.load(os.path.abspath(os.path.join(os.getcwd(), 'dn_train_x2_bi.npy')))
dn_test_bi_x2 = np.load(os.path.abspath(os.path.join(os.getcwd(), 'dn_test_x2_bi.npy')))
dn_train_bi_x4 = np.load(os.path.abspath(os.path.join(os.getcwd(), 'dn_train_x4_bi.npy')))
dn_test_bi_x4 = np.load(os.path.abspath(os.path.join(os.getcwd(), 'dn_test_x4_bi.npy')))
```

```python
plt.figure(figsize = (16,8))
plt.subplot(1,2,1)
plt.title('Training Error for DenseNet-DBPN-Pipeline')
plt.xlabel('Epochs')
plt.ylabel('Error %')
dn_train_h = plt.plot(100-dn_train[:300], label='DenseNet-Base')
dn_train_dbpn_x2_h = plt.plot(100-dn_train_dbpn_x2[:300], label='DenseNet-DBPN-x2')
dn_train_dbpn_x4_h = plt.plot(100-dn_train_dbpn_x4[:300], label='DenseNet-DBPN-x4')
plt.legend(handles = [dn_train_h[0],dn_train_dbpn_x2_h[0],dn_train_dbpn_x4_h[0]])

plt.subplot(1,2,2)
plt.title('Training Error for DenseNet-Bicubic-Pipeline')
plt.xlabel('Epochs')
plt.ylabel('Error %')
dn_train_h = plt.plot(100-dn_train[:300], label='DenseNet-Base')
dn_train_bi_x2_h = plt.plot(100-dn_train_bi_x2[:300], label='DenseNet-BiCubic-x2')
dn_train_bi_x4_h = plt.plot(100-dn_train_bi_x4[:300], label='DenseNet-BiCubic-x4')
plt.legend(handles = [dn_train_h[0],dn_train_bi_x2_h[0],dn_train_bi_x4_h[0]])
plt.show()
```

```python
plt.figure(figsize = (16,8))
plt.subplot(1,2,1)
plt.title('Validation Error for DenseNet-DBPN-Pipeline')
plt.xlabel('Epochs')
plt.ylabel('Error %')
dn_test_h = plt.plot(100-dn_test[:300], label='DenseNet-Base')
dn_test_dbpn_x2_h = plt.plot(100-dn_test_dbpn_x2[:300], label='DenseNet-DBPN-x2')
dn_test_dbpn_x4_h = plt.plot(100-dn_test_dbpn_x4[:300], label='DenseNet-DBPN-x4')
plt.legend(handles = [dn_test_h[0],dn_test_dbpn_x2_h[0],dn_test_dbpn_x4_h[0]])

plt.subplot(1,2,2)
plt.title('Validation Error for DenseNet-Bicubic-Pipeline')
plt.xlabel('Epochs')
plt.ylabel('Error %')
dn_test_h = plt.plot(100-dn_test[:300], label='DenseNet-Base')
dn_test_bi_x2_h = plt.plot(100-dn_test_bi_x2[:300], label='DenseNet-BiCubic-x2')
dn_test_bi_x4_h = plt.plot(100-dn_test_bi_x4[:300], label='DenseNet-BiCubic-x4')
plt.legend(handles = [dn_test_h[0],dn_test_bi_x2_h[0],dn_test_bi_x4_h[0]])
plt.show()
```

```python
#table results:
print('network | training error | testing error')
print('DenseNet-Base | ', np.min(100-dn_train), ' | ', np.min(100-dn_test))
print('DenseNet-BiCubic-x2 | ', np.min(100-dn_train_bi_x2), ' | ', np.min(100-dn_test_bi_x2))
print('DenseNet-BiCubic-x4 | ', np.min(100-dn_train_bi_x4), ' | ', np.min(100-dn_test_bi_x4))
print('DenseNet-DBPN-x2 | ', np.min(100-dn_train_dbpn_x2), ' | ', np.min(100-dn_test_dbpn_x2))
print('DenseNet-DBPN-x4 | ', np.min(100-dn_train_dbpn_x4), ' | ', np.min(100-dn_test_dbpn_x4))
```
